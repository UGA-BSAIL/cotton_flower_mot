```python
from cotton_flower_mot.pipelines.dataset_io import inputs_and_targets_from_datasets, HeatMapSource
from cotton_flower_mot.pipelines.config import ModelConfig

# Create the preprocessing pipeline.
config = ModelConfig(image_input_shape=(100, 100, 3), frame_input_shape=(1080, 1920, 3), detection_model_input_shape=(540, 960, 3),
                     rot_net_input_shape=(360, 360, 3), colorization_input_shape=(270, 480, 3),
                     colorization_output_shape=(72, 120, 32), num_appearance_features=128, num_node_features=128, num_edge_features=128, sinkhorn_lambda=10.0, num_reduction_stages=2,
                     detection_sigma=3, nominal_detection_size=(0.052, 0.093))
preprocessed = inputs_and_targets_from_datasets(raw_datasets=clips, config=config, heat_map_source=HeatMapSource.NONE, drop_probability=0.75, repeats=4)
```

```python
import itertools
from PIL import Image
from IPython.display import display

# Sample the data.
for inputs, targets in itertools.islice(preprocessed.unbatch(), 10):
    print("Detections:")
    for image in inputs["detections"]:
        # Convert BGR to RGB.
        image = Image.fromarray(image.numpy()[..., ::-1])
        display(image)
    
    print("Tracklets:")
    for image in inputs["tracklets"]:
        image = Image.fromarray(image.numpy()[..., ::-1])
        display(image)
        
    print(f"Sinkhorn:\n{targets['sinkhorn'].numpy()}")
```
